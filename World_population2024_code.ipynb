```python
import pandas as pd
import geopandas as gpd

data = pd.read_html('https://www.worldometers.info/world-population/')
pop_by_country = data[4]
print(pop_by_country)
```

```python
#for i in data:
 #   print(i)
    
pop_by_country = pop_by_country[['Country (or dependency)', 'Population (2024)' ]]
pop_by_country = pop_by_country.rename(columns={'Country (or dependency)': 'NAME'})
pop_by_country = pop_by_country.rename(columns={'Population (2024)': 'Population'})
print(pop_by_country.head())
```

```python
world_data = gpd.read_file(r"D:\JP CERTIFICATES\IITK\Course summerterm\AArcGIS project\python_integration\Coronavirus Cases March 2020\Coronavirus Cases March 2020\World_Map.shp")


world_data.plot(figsize=(10,10))
```

```python
for items in pop_by_country['NAME'].tolist():
    world_data_list = world_data['NAME'].tolist()
    if items in world_data_list:
        pass
    else:
        print(items + ' is not in the world_data list')
    
#print(world_data_list)
```

```python
world_data.replace('Korea, Republic of', 'S. Korea', inplace = True)
world_data.replace('Iran (Islamic Republic of)', 'Iran', inplace = True)
world_data.replace('United States', 'USA', inplace = True)
world_data.replace('United Kingdom', 'U.K.', inplace = True)
world_data.replace('United Arab Emirates', 'U.A.E.', inplace = True)
world_data.replace('Viet Nam', 'Vietnam', inplace = True)
world_data.replace('Macau', 'Macao', inplace = True)
world_data.replace('The former Yugoslav Republic of Macedonia', 'North Macedonia', inplace = True)
world_data.replace('Czech Republic', 'Czechia', inplace = True)
world_data.replace('Czech Republic', 'Czechia', inplace = True)
world_data.replace('Palestine', 'State of Palestine', inplace = True)
```

```python
combined = world_data.merge(pop_by_country, on = 'NAME')

combined.to_file(r'D:\JP CERTIFICATES\IITK\Course summerterm\AArcGIS project\python_integration\Coronavirus Cases March 2020\Coronavirus Cases March 2020\combined.shp')
```
